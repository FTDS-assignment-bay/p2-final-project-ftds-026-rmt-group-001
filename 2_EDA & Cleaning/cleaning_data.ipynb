# Fill the missing value

```python
#for object column

fill_value = "Tidak ada keterangan"

df['Sertifikat'] = df['Sertifikat'].fillna(fill_value)
df['Daya_Listrik'] = df['Daya_Listrik'].fillna(fill_value)
df['Ruang_Makan'] = df['Ruang_Makan'].fillna(fill_value)
df['Ruang_Tamu'] = df['Ruang_Tamu'].fillna(fill_value)
df['Kondisi_Perabotan'] = df['Kondisi_Perabotan'].fillna(fill_value)
df['Pemandangan'] = df['Pemandangan'].fillna(fill_value)
df['Terjangkau_Internet'] = df['Terjangkau_Internet'].fillna(fill_value)
df['Lebar_Jalan'] = df['Lebar_Jalan'].fillna(fill_value)
df['Sumber_Air'] = df['Sumber_Air'].fillna(fill_value)
df['Hook'] = df['Hook'].fillna(fill_value)
df['Kondisi_Properti'] = df['Kondisi_Properti'].fillna(fill_value)
```

```python
#for numeric column

fill_value = 0

df['Kamar_Tidur'] = df['Kamar_Tidur'].fillna(fill_value)
df['Kamar_Mandi'] = df['Kamar_Mandi'].fillna(fill_value)
df['Carport'] = df['Carport'].fillna(fill_value)
df['Kamar_Pembantu'] = df['Kamar_Pembantu'].fillna(fill_value)
df['Dapur'] = df['Dapur'].fillna(fill_value)
df['Jumlah_Lantai'] = df['Jumlah_Lantai'].fillna(fill_value)
```
