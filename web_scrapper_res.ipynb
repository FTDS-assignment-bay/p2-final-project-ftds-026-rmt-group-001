```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome()

url="https://www.rumah123.com/jual/balikpapan/residensial/"
driver.get(url)
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")
print(soup.prettify()[:700])
```

```python
judul = []
hyperlink = []
img_hyperlinks = []
prices = [] 
info = [] 
kategori = []
driver = webdriver.Chrome()

# looping function to traverse the website
for i in range(1, 35):
    url = "https://www.rumah123.com/jual/residensial/?q=Balikpapan&page=" + str(i)
    print(url)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # get the id of the house listing
    get_id = soup.find_all('div', class_="card-featured__middle-section")

    # inside div get the hyperlink and price of the house listing
    for i in get_id:
        get_hyperlink = i.find('a', target="_blank")
        if get_hyperlink:
            hyperlink.append("https://www.rumah123.com/" + get_hyperlink.get('href'))
            # print("https://www.rumah123.com/" + get_hyperlink.get('href'))

            # get the title from the 'a' tag
            judul.append(get_hyperlink.get('title'))

            # get the price from the listing
            price_tag = i.find('div', class_="card-featured__middle-section__price")
            if price_tag:
                price = price_tag.text.strip()
                prices.append(price)
                # print(price)

            # get the info from the listing
            lb_tag = i.find('div', class_="card-featured__middle-section__attribute")
            if lb_tag:
                lb = lb_tag.text.strip()
                info.append(lb)
                # print(info)

            # get the info from the listing
            cat_tag = i.find('div', class_="card-featured__middle-section__header-badge")
            if cat_tag:
                cat = cat_tag.text.strip()
                kategori.append(cat)
                # print(category)
            
    get_img = soup.find_all('div', class_="card-featured__media-section card-featured__shimmer")
    # iterate over the divs
    for div in get_img:
        # get the image hyperlink
        img = div.find('img')
        if img:
            img_hyperlink = img.get('src')
            img_hyperlinks.append(img_hyperlink)
            # print(img_hyperlink)

# Close the webdriver when done
driver.quit()
```

```python
# save hyperlink and judul to dataframe
df_link = pd.DataFrame({'judul':judul, 'kategori':kategori, 'prices':prices, 'info':info, 'hyperlink':hyperlink, 'img_hyperlinks':img_hyperlinks})
df_link
```

```python
#Save table to csv
df_link.to_csv('rumah123_res_hyperlink.csv', index=False)
```
