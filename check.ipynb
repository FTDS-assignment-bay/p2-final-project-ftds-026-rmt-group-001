```python
# code to force show all column
pd.set_option('display.max_columns', None)

# max column width
pd.set_option('display.max_colwidth', 100)
```

```python
# load the csvs to join

csv_hyperlink = pd.read_csv('./datasets/rumah123_hyperlink.csv')
csv_data = pd.read_csv('./datasets/rumah123_data_single_raw.csv')
```

```python
# get 10 random rows from csv_data
random_rows = csv_data.sample(10)
random_rows
```

```python
# inner join csv data and hyperlink on Judul and judul

merged = pd.merge(csv_data, csv_hyperlink, left_on='Judul', right_on='judul', how='inner')
merged
```

```python
# count check for duplicates on merged
merged.duplicated().sum()

# drop duplicates
merged = merged.drop_duplicates()
```
