```python
import pandas as pd
from itertools import zip_longest
from bs4 import BeautifulSoup
from selenium import webdriver

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import time
```

```python
# code to force show all column
pd.set_option('display.max_columns', None)

# max column width
pd.set_option('display.max_colwidth', 300)
```

```python
# load all dataset
df1 = pd.read_csv('../datasets/rumah123_data_single_raw_concated_part_1.csv')
df2 = pd.read_csv('../datasets/rumah123_data_single_raw_concated_part_2.csv')
df3 = pd.read_csv('../datasets/rumah123_data_single_raw_concated_part3.csv')
```

```python
# concat all dataset
df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df
```

```python
# check duplicate
dup = df[df['Judul'].duplicated()]
```

```python
# rename the duplicated judul row to be unique
dup['Judul'] = dup['Judul'] + ' Seharga ' + dup['Harga'].astype(str)
```

```python
# replace the duplicated row with the new unique row
df = df.drop(dup.index)
df = pd.concat([df, dup], ignore_index=True)
```

```python
df[df['Judul'].duplicated()].sort_values('Judul')
```

```python
# to csv
df.to_csv('../datasets/rumah123_data_single_raw_concated.csv', index=False)
```
