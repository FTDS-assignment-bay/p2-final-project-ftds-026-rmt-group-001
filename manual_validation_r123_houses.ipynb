# Solution: Inner Join the two separate csv on 'Judul'

So the problem lies with how the scraped data and scraped hyperlinks are concatenated. There's different title bcs both of those tables have the field 'judul' and the concatenation was not on the same judul, but instead a basic concatenation.

judul from the hyperlink table will be used as inner join base

```python
# load the csvs to join

csv_hyperlink = pd.read_csv('./datasets/rumah123_hyperlink.csv')
csv_data = pd.read_csv('./datasets/rumah123_data_single_raw.csv')
```

```python
# look at column names, especially 'Judul'

csv_data.head()
```

```python
# look at column names, especially 'judul'

csv_hyperlink.head()
```

Insight : notice how in csv_hyperlink it's 'judul' while on csv_data it's 'Judul' with a capital J. If we want to join these two tables on 'judul', the name would have to be the same for the program to recognize them as same.

```python
# rename column 'Judul' in csv_data to 'judul'

csv_data.rename(columns={'Judul':"judul"}, inplace=True)

csv_data.head()
```

```python
csv_inner_join = csv_hyperlink.merge(csv_data, on='judul', how='inner')

csv_inner_join.tail()

```
